**75. How to compute averages using a sliding window over an array?**

```python
def moving_average(arr, n=3) :
    rearange = np.cumsum(arr)
    rearange[n:] = rearange[n:] - rearange[:-n]
    return rearange[n - 1:] / n

arr = np.arange(20)

print(moving_average(arr,3))
```
