```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

# load data and get the df

```python
data = load_breast_cancer(return_X_y=False) #If True, returns (data, target) instead of a Bunch object.
```

```python
def sklearn_to_df(sklearn_dataset):
    df = pd.DataFrame(sklearn_dataset.data, columns=sklearn_dataset.feature_names)
    df['target'] = pd.Series(sklearn_dataset.target)
    return df
```

```python
df = sklearn_to_df(data)
```

```python
df.head()
```

```python
df5 = df[['mean radius', 'mean texture','mean perimeter','mean area','mean smoothness']]
```

```python
display(df5.head())
print(len(df5))
```

# correlation

```python
corr_matrix = df5.corr()
corr_matrix
```

# conditional probability for two columns

```python
cp_radius_mean = pd.crosstab(index = df5['mean radius'], #Compute a simple cross tabulation of two (or more) factors.
           columns = df5['mean texture'],
           margins = True,   # add = subtotals
           dropna = True,
           normalize = 'index') #row normalizing so row sum is 1
display(cp_radius_mean)
```

```python
print(cp_radius_mean.sum(axis=1)) #check if all row is in sum 1 
```

```python
cp_radius_mean_col = pd.crosstab(index = df5['mean radius'],
          columns = df5['mean texture'],
          margins = True,
          dropna = True,
          normalize = 'columns') #col normalizing so col sum is 1
cp_radius_mean_col
```

```python
print(cp_radius_mean_col.sum(axis=0)) #check if all col is in sum 1 
```

# conditional probability with mutliple columns

```python
conditional_prob= pd.crosstab(index = df5['mean radius'], #Compute a simple cross tabulation of two (or more) factors.
           columns = [df5['mean texture'],df5['mean perimeter'],df5['mean area'],df5['mean smoothness']],
           margins = True,   # add = subtotals
           dropna = True,
           normalize = 'index') #row normalizing so row sum is 1
display(conditional_prob)
```

```python
print(conditional_prob.sum(axis=1)) #check if all row is in sum 1 
print(conditional_prob.sum(axis=0)) #check if all col is in sum 1 
```

```python
conditional_prob_col= pd.crosstab(index = df5['mean radius'], #Compute a simple cross tabulation of two (or more) factors.
           columns = [df5['mean texture'],df5['mean perimeter'],df5['mean area'],df5['mean smoothness']],
           margins = True,   # add = subtotals
           dropna = True,
           normalize = 'columns') #row normalizing so row sum is 1 ##### doesnt work with multiple columns ??
display(conditional_prob_col)
```

```python
print(conditional_prob_col.sum(axis=1)) #check if all row is in sum 1 
print(conditional_prob_col.sum(axis=0)) #check if all col is in sum 1 
```
