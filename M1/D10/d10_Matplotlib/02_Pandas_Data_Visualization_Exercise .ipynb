This is just a quick exercise for you to review the various plots we showed earlier. Use **df3** to replicate the following plots. 

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
data = {"a": np.random.random(500),
        "b": np.random.random(500),
        "c": np.random.random(500),
        "d": np.random.random(500)}
df3 = pd.DataFrame(data)
%matplotlib inline
```

**Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion. Remeber back to your matplotlib lecture...**

```python
# thicks inside on top and right ?
fig, ax = plt.subplots(figsize=(12,3))
plt.scatter(df3['a'],df3['b'],
            color='red',
            edgecolors='black',
           s= 50) #scatter(x,y)
plt.ylim(-0.2,1.2)
plt.xlim(-0.2,1.2)
plt.xlabel('a')
plt.ylabel('b')
plt.tick_params(axis='both',direction='in',top=True,right=True)
plt.show()
```

**Create a histogram of the 'a' column.**

```python

fig, ax = plt.subplots(figsize=(7,5))
plt.hist(df3['a'],color='blue',edgecolor='black')
plt.xlim(0.0,1.0)
plt.ylabel('Frequency')
plt.tick_params(axis='both',direction='in',top=True,right=True)
plt.show()
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(7,5))
plt.hist(df3['a'],bins = 25,color='salmon', alpha=0.7,edgecolor='white')
plt.xlim(0.0,1.0)
plt.ylabel('Frequency')
plt.tick_params(axis='both',direction='out',top=True,right=True)
plt.show()
```

```python
df = [df3['a'],df3['b']]
fig, ax = plt.subplots(figsize=(10,8))
box = plt.boxplot(df,
                  boxprops = dict(
                      color='r'),
                  whiskerprops = dict(
                      linestyle='--',
                      linewidth=2,
                      color='r'),
                 medianprops = dict(
                     linestyle='-',
                     color = 'b'))
#for _, line_list in box.items():
#    for line in line_list:
#        line.set_color('r')
ax.set_xticklabels(['a', 'b'])
plt.tick_params(axis='both',direction='out',top=True,right=True)
plt.ylim(0.0,1.0)
```

**Create a kde plot of the 'd' column**

```python
df3['d'].plot.kde()
plt.ylim(0.0,1.2)
plt.xlim(-.5,1.5)
plt.xticks(np.arange(-0.5,1.6, 0.5))
plt.tick_params(axis='both',direction='out',top=True,right=True)
plt.show()
```
